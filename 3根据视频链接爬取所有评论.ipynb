# 爬取视频评论文本及时间

## 导入需要的库

```python
import requests
import re
import json
import time
import csv
```

## 根据视频链接获取视频的aid，后续爬取评论的时候需要aid

```python
def getVideoAid(url):
    ret = requests.get(url)
    aid = re.findall('\"aid\":(.*?),',ret.text)[0]
    return aid
```

## 爬取评论内容

```python
def getComments(oid,page):
    url = 'https://api.bilibili.com/x/v2/reply?jsonp=jsonp&pn={}&type=1&oid={}&sort=2'.format(page,oid)
    ret = requests.get(url)
    return ret.text
```

## 保存评论数据

```python
def saveComments(data_list):
    f = open('comments.csv','a',encoding='utf-8-sig',newline='')
    c = csv.writer(f)
    for data in data_list:
        c.writerow(data)
    f.close()
```

## 解析评论数据并返回当前视频的评论页数

```python
def parseComments(htmlData):
    try:
        resultData = json.loads(htmlData)['data']
        pageData = int(resultData['page']['count'] / 20) + 1
        data_list = resultData['replies']
        saveList = []
        for data in data_list:
            ctime = data['ctime']  # 评论时间
            ctime_str = time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(ctime))
            comment = data['content']['message']
            saveList.append([comment,ctime_str])
        saveComments(saveList)
        return pageData
    except:
        return 1
```

## 控制爬虫逻辑的函数

```python
def controlSpider():
    f = open('18videos.txt', 'r')
    textList = f.readlines()
    f.close()
    for k, url in enumerate(textList):
        try:
            oid = getVideoAid(url.replace('\n', ''))
        except:
            continue
        time.sleep(5)
        page = 1
        new_page = 1
        while page <= new_page:
            time.sleep(4)
            html_data = getComments(oid,page)
            new_page = parseComments(html_data)
            print('comments get success page:%s!'%page)
            page += 1
```

## 爬取数据
