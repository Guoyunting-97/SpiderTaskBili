# 爬虫B站美食搜索mukbang

## 导入所需要的库

```python
import requests
import time
import csv
import json
```

## 获取网页源码函数

```python
def getHtml(page):
    url = 'https://api.bilibili.com/x/web-interface/search/type?context=&page={}' \
          '&order=dm&keyword=mukbang&duration=0&tids_2=&__refresh__=true&_extra=' \
          '&search_type=video&tids=0&highlight=1&single_column=0'.format(page)
    ret =requests.get(url)
    return ret.text
```

## 保存数据函数

```python
def saveData(dataList):
    print(dataList)
    f = open('result.csv','a',newline='',encoding='utf-8-sig')
    c = csv.writer(f)
    for data in dataList:
        c.writerow(data)
    f.close()
```

## 从网页源码中提取数据函数

```python
def parseIndexHtml(htmlData):
    dataList = json.loads(htmlData)['data']['result']
    myData = []
    for data in dataList:
        author = data['author']
        # mid = data['mid']
        # aid = data['aid']
        bvid = data['bvid']
        url = 'https://www.bilibili.com/video/{}?from=search'.format(bvid)
        title = data['title']
        play = data['play']  # 播放量
        duration = data['duration']  # 视频时长
        typename = data['typename']
        tag = data['tag']
        if typename not in tag:
            tag += ',' + typename
        myData.append([author,title,tag,play,duration,url])
    saveData(myData)
```

## 控制爬虫逻辑函数

```python
def spiderRangePage():
    f = open('result.csv', 'w', newline='', encoding='utf-8-sig')
    c = csv.writer(f)
    c.writerow([
        '作者',
        '标题',
        '关键词',
        '播放量',
        '视频时长',
        '链接'
    ])
    f.close()
    for page in range(50):
        htmlData = getHtml(page+1)
        parseIndexHtml(htmlData)
        time.sleep(3)
```

## 运行爬虫

```python
spiderRangePage()
```
